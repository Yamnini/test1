```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Name': [ 'john', 'vishan', 'kishan', 'arthi'],
    'Maths': [85, 75, 96, 87],
    'Science': [95, 55, 56, 80],
    'Social': [50, 85, 89, 60],
    'English': [65, 45, 89, 65]
}
df = pd.DataFrame(data)
pivot = df.melt(id_vars='Name', var_name='Subject', value_name='Score')
print(pivot)
new_row={"Name":"New Student","Maths":90,"Science":70,"Social":60,"English":80}
df=df._append(new_row,ignore_index=True)
df["New Subject"]=[100,90,80,70,60]
sorted_df = df.sort_values(by='Maths', ascending=False)
print(sorted_df)
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Maths'], label='Maths')
plt.bar(df['Name'], df['Science'], label='Science')
plt.bar(df['Name'], df['Social'], label='Social')
plt.bar(df['Name'], df['English'], label='English')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df['Name'], df['Maths'], marker='o', label='Maths')
plt.plot(df['Name'], df['Science'], marker='o', label='Science')
plt.plot(df['Name'], df['Social'], marker='o', label='Social')
plt.plot(df['Name'], df['English'], marker='o', label='English')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Ename': ['Roshan', 'Amar', 'Ashwini', 'Lohith', 'Mohan', 'Pramod'],
    'Type': ['Regular', 'Adhoc', 'Regular', 'Adhoc', 'Contract', 'Regular'],
    'Experience': [10, 20, 5, 14, 9, 8],
    'Salary': [50000, 15000, 30000, 15000, 10000, 40000],
    'Department': ['CS', 'CS', 'EC', 'EC', 'CS', 'EC']
}
df = pd.DataFrame(data)
pivot1 = df.pivot_table(index='Ename', columns='Department', values='Salary', aggfunc='mean')
print(pivot1)
pivot2 = df.pivot_table(index='Type', values='Salary', aggfunc=['sum', 'mean'])
pivot2['Count'] = df['Type'].value_counts()
print(pivot2)
plt.figure(figsize=(10, 6))
plt.bar(df['Type'], df['Salary'])
plt.xlabel('Type')
plt.ylabel('Salary')
plt.title('Salary by Type')
plt.show()

plt.figure(figsize=(10, 6))
plt.pie(df['Type'].value_counts(), labels=df['Type'].unique(), autopct='%1.1f%%')
plt.title('Employee Type Distribution')
plt.show()
```
